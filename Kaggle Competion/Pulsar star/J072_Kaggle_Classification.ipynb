```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
from matplotlib import style

import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
train.shape
```

```python
train.info()
```

```python
train.describe(include='all')
```

```python
train.isnull().sum()
```

```python
X = train[['Mean of the integrated profile','Standard deviation of the integrated profile','Excess kurtosis of the integrated profile','Skewness of the integrated profile','Mean of the DM-SNR curve','Standard deviation of the DM-SNR curve','Excess kurtosis of the DM-SNR curve','Skewness of the DM-SNR curve']]
y = train[['target_class']]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
from sklearn import metrics
from sklearn.svm import LinearSVC
model = LinearSVC()
```

```python
reg = model.fit(X_train, y_train)
reg.score(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred)
```
